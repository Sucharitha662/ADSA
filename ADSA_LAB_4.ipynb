```python
#Towers of Hanoi Using Recursive Approach
def towers_of_hanoi(n, source, auxiliary, destination):
    if n == 1:
        print(f"Move disk 1 from {source} → {destination}")
        return
    # Move n-1 disks from source to auxiliary
    towers_of_hanoi(n - 1, source, destination, auxiliary)
    # Move the nth disk from source to destination
    print(f"Move disk {n} from {source} → {destination}")
    # Move the n-1 disks from auxiliary to destination
    towers_of_hanoi(n - 1, auxiliary, source, destination)


# --- Main Program ---
num_disks = int(input("Enter the number of disks: "))
print(f"\nSteps to solve Tower of Hanoi for {num_disks} disks:\n")
towers_of_hanoi(num_disks, 'A', 'B', 'C')
```

```python
#Towers of Hanoi Using Stack Based Iteration
# Stack-based Iterative Tower of Hanoi Solution

class Move:
    def __init__(self, n, source, auxiliary, destination, state=0):
        self.n = n
        self.source = source
        self.auxiliary = auxiliary
        self.destination = destination
        self.state = state  # Tracks which part of the recursive step we are in


def tower_of_hanoi_iterative(n, source, auxiliary, destination):
    stack = []
    stack.append(Move(n, source, auxiliary, destination))
    moves = []

    while stack:
        top = stack.pop()

        # If only one disk, perform move directly
        if top.n == 1:
            moves.append(f"Move disk 1 from {top.source} → {top.destination}")
        else:
            # Simulate recursion using state tracking
            if top.state == 0:
                # Step 1: Move n-1 disks to auxiliary
                top.state = 1
                stack.append(top)
                stack.append(Move(top.n - 1, top.source, top.destination, top.auxiliary))
            elif top.state == 1:
                # Step 2: Move the largest disk
                moves.append(f"Move disk {top.n} from {top.source} → {top.destination}")
                # Step 3: Move n-1 disks from auxiliary to destination
                stack.append(Move(top.n - 1, top.auxiliary, top.source, top.destination))

    return moves


# --- Main Program ---
n = int(input("Enter the number of disks: "))
print(f"\nSteps to solve Tower of Hanoi for {n} disks (Iterative using Stack):\n")
result = tower_of_hanoi_iterative(n, 'A', 'B', 'C')

# Print all moves
for move in result:
    print(move)
```
