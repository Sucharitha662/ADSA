```python
#Create an Undo-Redo System Using Two Stacks. Implement Push, Pop, and Display Operations.
class UndoRedoSystem:
    def __init__(self):
        self.undo_stack = []
        self.redo_stack = []

    def push(self, action):
        """Push operation — add a new action"""
        self.undo_stack.append(action)
        self.redo_stack.clear()  # clear redo stack since new action added
        print(f"Action '{action}' pushed successfully.")

    def pop(self):
        """Pop operation — Undo the last action"""
        if not self.undo_stack:
            print("No actions to undo!")
            return
        action = self.undo_stack.pop()
        self.redo_stack.append(action)
        print(f"Undo performed: '{action}' removed.")

    def redo(self):
        """Redo operation — Redo the last undone action"""
        if not self.redo_stack:
            print("No actions to redo!")
            return
        action = self.redo_stack.pop()
        self.undo_stack.append(action)
        print(f"Redo performed: '{action}' re-applied.")

    def display(self):
        """Display the current contents of both stacks"""
        print("\n------ STACK STATUS ------")
        print("Undo Stack:", self.undo_stack if self.undo_stack else "Empty")
        print("Redo Stack:", self.redo_stack if self.redo_stack else "Empty")
        print("--------------------------\n")


# ---------- Main Menu ----------
if __name__ == "__main__":
    system = UndoRedoSystem()

    while True:
        print("1. Push Operation")
        print("2. Pop (Undo) Operation")
        print("3. Redo Operation")
        print("4. Display Stacks")
        print("5. Exit")

        try:
            choice = int(input("Enter your choice: "))
        except ValueError:
            print("Invalid input! Please enter a number.\n")
            continue

        if choice == 1:
            action = input("Enter the action to push: ")
            system.push(action)
        elif choice == 2:
            system.pop()
        elif choice == 3:
            system.redo()
        elif choice == 4:
            system.display()
        elif choice == 5:
            print("Exiting program...")
            break
        else:
            print("Invalid choice! Please enter between 1 and 5.\n")
```
