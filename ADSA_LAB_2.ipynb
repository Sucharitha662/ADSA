```python
#Convert The Patient List to A Circular Linked List for a Round-Robin Check-Up System. Implement Insertion and Deletion.

import time  # to simulate time delay for check-up rounds

class PatientNode:
    def __init__(self, name, age, patient_id):
        self.name = name
        self.age = age
        self.patient_id = patient_id
        self.next = None


class CircularPatientList:
    def __init__(self):
        self.head = None

    def insert_patient(self, name, age, patient_id):
        new_node = PatientNode(name, age, patient_id)
        if self.head is None:
            self.head = new_node
            new_node.next = self.head
        else:
            current = self.head
            while current.next != self.head:
                current = current.next
            current.next = new_node
            new_node.next = self.head
        print(f"Inserted: {name}, Age: {age}, ID: {patient_id}")

    def delete_patient(self, patient_id):
        if self.head is None:
            print("No patients to delete.")
            return

        current = self.head
        previous = None

        while True:
            if current.patient_id == patient_id:
                if previous is not None:
                    previous.next = current.next
                else:
                    # Deleting the head node
                    temp = self.head
                    while temp.next != self.head:
                        temp = temp.next
                    if self.head == self.head.next:
                        self.head = None
                        print(f"Deleted patient with ID: {patient_id}")
                        return
                    temp.next = self.head.next
                    self.head = self.head.next
                print(f"Deleted patient with ID: {patient_id}")
                return
            previous = current
            current = current.next
            if current == self.head:
                break

        print(f"Patient with ID {patient_id} not found.")

    def display_patients(self):
        if self.head is None:
            print("No patients in the list.")
            return

        print("\nPatient List (Round-Robin Order):")
        current = self.head
        while True:
            print(f"Name: {current.name}, Age: {current.age}, ID: {current.patient_id}")
            current = current.next
            if current == self.head:
                break

    def simulate_checkups(self, rounds=2, delay=1.5):
        """Simulate round-robin patient check-ups."""
        if self.head is None:
            print("No patients to check.")
            return

        print("\n Starting Round-Robin Check-Up Simulation...\n")
        current = self.head
        for round_num in range(1, rounds + 1):
            print(f"--- Round {round_num} ---")
            while True:
                print(f"Now checking: {current.name} (Age: {current.age}, ID: {current.patient_id})")
                time.sleep(delay)  # simulate time gap between checkups
                current = current.next
                if current == self.head:
                    break
            print()
        print("✅ Check-up simulation completed.\n")


# Example usage
if __name__ == "__main__":
    plist = CircularPatientList()

    # Insert 5 patients (custom order)
    plist.insert_patient("Alice", 30, 101)
    plist.insert_patient("Charlie", 25, 102)
    plist.insert_patient("Daniel", 50, 103)
    plist.insert_patient("Bob", 45, 104)
    plist.insert_patient("Fiona", 29, 105)

    # Display patients
    plist.display_patients()

    # Delete two patients
    plist.delete_patient(102)  # Charlie
    plist.delete_patient(101)  # Alice

    # Display updated list
    plist.display_patients()

    # Simulate Round-Robin Check-Up for remaining patients
    plist.simulate_checkups(rounds=3, delay=1)
```
